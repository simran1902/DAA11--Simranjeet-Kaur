```python
import pandas as pd
import numpy as np
import corner
import matplotlib.pyplot as plt

filenames = [
    "ACV.csv", "CEP.csv", "DCEP.csv"
]

names = ['Amplitude', 'AndersonDarling', 'Autocor_length', 'Beyond1Std', 'CAR_mean', 'CAR_sigma', 'CAR_tau', 'Con', 'Eta_e',
         'FluxPercentileRatioMid20', 'FluxPercentileRatioMid35']

# Initialize an empty list to collect all samples
all_samples = []

# Function to process each file and create corner plots
def process_file(filename):
    data_in = pd.read_csv(filename, sep=',', names=names, encoding='unicode_escape')
    lc = data_in.to_records(index=False)
    
    # Collect data for all names
    sample = []
    for name in names:
        sample.append(lc[name][1:].astype(float))
    
    return np.array(sample).T

# Define a list of colors
colors = plt.cm.tab20(np.linspace(0, 1, len(filenames)))

# Loop through each file and generate plots
for i, filename in enumerate(filenames):
    sample = process_file(filename)
    all_samples.append((sample, colors[i]))
    print(f"Sample size for {filename}: {len(sample)}")

# Create a single corner plot with all samples
figure = None
for sample, color in all_samples:
    figure = corner.corner(sample, labels=names, color=color, 
                           quantiles=[0.16, 0.5, 0.84], show_titles=True, 
                           title_kwargs={"fontsize": 12}, fig=figure)

figure.suptitle('Combined Corner Plot for All Datasets', fontsize=16)
plt.show()
figure.savefig("combined_corner_plot.png")

print("Combined corner plot generated.")
```
